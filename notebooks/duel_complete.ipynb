## IMPORTANDO AS BIBLIOTECAS NECESSÁRIAS

# SELECIONANDO COMPETIÇÃO E TEMPORADA

```python
df_championship = sb.matches(competition_id=43 , season_id=106 )
```

```python
# Checando as equipes que disputaram a competição
df_championship['home_team'].sort_values().unique()
```

```python
def data(team):
    
    df_team = df_championship.loc[(df_championship['home_team'] == f'{team}')|(df_championship['away_team'] == f'{team}')]
    match_list = df_team['match_id'].unique().tolist()
        
    # Inicializar o parser
    parser = Sbopen()

    # Listas para armazenar os dataframes
    all_df = []
    all_related = []
    all_freeze = []
    all_tactics = []

    # Iterar sobre cada ID de jogo e coletar os dados
    for id in match_list:
        df, related, freeze, tactics = parser.event(id)
            
        # Adicionar os dataframes coletados às listas correspondentes
        all_df.append(df)
        all_related.append(related)
        all_freeze.append(freeze)
        all_tactics.append(tactics)

    df_combined = pd.concat(all_df, ignore_index=True) # Informações de eventos detalhados
    related_combined = pd.concat(all_related, ignore_index=True) # Informação de eventos relacionados (ex: Pass - Ball Receipt)
    freeze_combined = pd.concat(all_freeze, ignore_index=True) # Informações de coordenada dos eventos
    tactics_combined = pd.concat(all_tactics, ignore_index=True) # informações dos jogadores(nº camisa, posição, nome e ID)
    
    return df_combined, related_combined, freeze_combined, tactics_combined, df_team
```

## SELECIONANDO EQUIPE

```python
# Escolha a equipe que deseja analisar e substitua na variável 'team'
team = 'Brazil'
df_combined, related_combined, freeze_combined, tactics_combined, df_team = data(team=team)
```

```python
# Extraindo os nomes completos dos jogadores da equipe selecionada anteriormente
df_players = df_combined[df_combined['team_name'] == team]
players_names = df_players['player_name'].sort_values().unique().tolist()
players_names
```

```python
# Extraindo a fração para facilitar a visualização do id do jogo para seleção
team_match_ids = df_team['match_id'].unique().tolist()
df_team[df_team['match_id'].isin(team_match_ids)][['match_id', 'home_team', 'away_team']]
```

```python
def duel_analysis(player=None, match=None):
    df_duel = df_combined[df_combined['type_name'] == 'Duel']
    df_duel = df_duel[df_duel['team_name'] == team]

    # Filtra duelos de acordo com o jogo (se fornecido)
    if match is not None:
        df_duel_filtered_match = df_duel[df_duel.match_id == match]
    else:
        df_duel_filtered_match = df_duel[df_duel.match_id.isin(team_match_ids)]  # Todos os jogos da equipe

    # Filtra duelos de acordo com o jogador (se fornecido)
    if player is not None:
        df_duel_filtered_player = df_duel_filtered_match[df_duel_filtered_match.player_name == player].reset_index(drop=True)
    else:
        df_duel_filtered_player = df_duel_filtered_match.reset_index(drop=True)  # Todos os jogadores

    # Plotar apenas se houver dados
    if df_duel_filtered_player.empty:
        print("Nenhum duelo encontrado para os filtros aplicados.")
        return

    start_x = df_duel_filtered_player.x
    start_y = df_duel_filtered_player.y
    outcome_name = df_duel_filtered_player.outcome_name


    # Cores para o outcome
    color_map = {
        'Won': '#0000FF',  # Azul para "Duelo Vencido e posse mantida"
        'Success In Play': '#00FF00',  # Verde para "Posse para outro jogador da própria equipe"
        'Success Out': '#FF0000',  # Vermelho para "Reposição para a própria equipe"
        'Lost': '#FFFF00',  # Amarelo para "Duelo Perdido"
        'Lost In Play': '#FFA500',  # Laranja para "Posse para outro jogador da equipe adversária"
        'Lost Out': '#800080'}  # Roxo para "Reposição para equipe adversária"

    # Define cor padrão para NaN e cria a lista de cores
    colors = [color_map.get(outcome, '#A9A9A9') if pd.notna(outcome) else '#A9A9A9' for outcome in outcome_name]

    # Campo
    pitch = Pitch(pitch_color='#aabb97', line_color='white', stripe_color='#c2d59d', stripe=True)
    fig, ax = pitch.draw(figsize=(10, 6))

    # Plotando os duelos
    for i in range(len(df_duel_filtered_player)):
        pitch.scatter(start_x[i], start_y[i], ax=ax, color=colors[i], edgecolor='black', alpha=0.7)


    # Legenda de resultados do duelo (outcome)
    legend_colors = [
        mpatches.Patch(color='#0000FF', label='Duelo Vencido e posse mantida'),
        mpatches.Patch(color='#00FF00', label='Posse para outro jogador da própria equipe'),
        mpatches.Patch(color='#FF0000', label='Reposição para a própria equipe'),
        mpatches.Patch(color='#FFFF00', label='Duelo Perdido'),
        mpatches.Patch(color='#FFA500', label='Posse para outro jogador da equipe adversária'),
        mpatches.Patch(color='#800080', label='Reposição para equipe adversária'),
        mpatches.Patch(color='#A9A9A9', label='Duelo Aéreo Perdido')]

    plt.legend(
        handles=legend_colors,
        loc='upper left',
        bbox_to_anchor=(1, 1),
        title="Legenda",
        title_fontsize=14,
        fontsize=10)

    # Título com base nas condições
    if match is not None:
        df_home_away = df_team[df_team['match_id'] == match]
        home = df_home_away['home_team'].iloc[0]
        away = df_home_away['away_team'].iloc[0]
        title_match_info = f"{home} Vs {away}"
    else:
        title_match_info = "Todos os jogos"

    if player is not None:
        title_player_info = f"{player}"
    else:
        title_player_info = f"{team}"

    plt.title(f'Análise de Duelos: {title_player_info} - {title_match_info} - {df_championship["competition"].loc[0]} {df_championship["season"].loc[0]}')
    plt.show()
```

## ANALISANDO

```python
# Para analisar todos os jogos da equipe na competição selecionada basta chamar a função sem nenhum argumento: ...analysis()
# Para analisar um jogo específico da equipe selecionada: ...analysis(match=match_id)
# Para analisar um jogador específico na competição inteira: ...analysis(player='player_name')
# Para analisar um jogador específico em um jogo específico: ...analysis(player='player_name', match=match_id)

duel_analysis(player='Vinícius José Paixão de Oliveira Júnior')
```
